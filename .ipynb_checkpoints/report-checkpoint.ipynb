## Finding Camera Calibration Matrix
    To solve for the projection matrix. We need to set up a system of
    equations using the corresponding 2D and 3D points:

                                                      [ M11      [ u1
                                                        M12        v1
                                                        M13        .
                                                        M14        .
    [ X1 Y1 Z1 1 0  0  0  0 -u1*X1 -u1*Y1 -u1*Z1        M21        .
      0  0  0  0 X1 Y1 Z1 1 -v1*X1 -v1*Y1 -v1*Z1        M22        .
      .  .  .  . .  .  .  .    .     .      .       *   M23   =    .
      Xn Yn Zn 1 0  0  0  0 -un*Xn -un*Yn -un*Zn        M24        .
      0  0  0  0 Xn Yn Zn 1 -vn*Xn -vn*Yn -vn*Zn ]      M31        .
                                                        M32        un
                                                        M33 ]      vn ]

    Then we can solve this using SVD with np.linalg.svd().
    We obtain 2 equations for each corresponding 2D and 3D point pair.Hence To solve this, we need at least 6 point pairs.

```python
def calculate_projection_matrix(points_2d, points_3d):
    n_p = points_3d.shape[0]
    A = []
    for i in range(n_p):
        x,y,z = points_3d[i,0] ,points_3d[i,1] , points_3d[i,2]
        u,v = points_2d[i,0] , points_2d[i,1]
        A.append([x,y,z,1,0,0,0,0,-u*x,-u*y,-u*z,-u])
        A.append([0,0,0,0,x,y,z,1,-v*x,-v*y,-v*z,-v])
    A = np.asarray(A)
    u,s,vh =  np.linalg.svd(A)
    L = vh[-1,:]/vh[-1,-1]
    M = L.reshape(3, 4)

    return M
```

```python
def norm(arr):
    arr = arr - np.mean(arr,axis=0 , keepdims = True)
    arr = arr / np.abs(arr).max(axis=0)
    return arr
```

### Evaluation Metrics
Following function calculates the new projection points using camera matrix M and real world points , and 
also the residual ( mean square error)


```python
def evaluate_points(M, points_2d, points_3d):
    N = len(points_3d)
    points_3d = np.hstack((points_3d, np.ones((N, 1))))
    points_3d_proj = np.dot(M, points_3d.T).T
    u = points_3d_proj[:, 0] / points_3d_proj[:, 2]
    v = points_3d_proj[:, 1] / points_3d_proj[:, 2]
    residual = np.sum(np.hypot(u-points_2d[:, 0], v-points_2d[:, 1]))/N
    points_3d_proj = np.hstack((u[:, np.newaxis], v[:, np.newaxis]))
    return points_3d_proj, residual
```

### Calculate the projection matrix given corresponding 2D & 3D points


```python
temp = np.concatenate((points_2d , points_3d), axis = 1)
np.random.shuffle(temp)
p2 = temp[:6,[0,1]]
p3 = temp[:6,[2,3,4]]
M = calculate_projection_matrix(p2, p3)
print('The projection matrix is\n')
print(M)
projected_2d_pts, residual = evaluate_points(M, points_2d, points_3d);

print('The total residual is {:f}'.format(residual))
```

# Calculating camera matrix using RANSAC

The idea is very simple: two of the points are selected randomly; these points define a line. The support for this line is measured by the number of points that lie within a distance threshold. This random selection is repeated a number of times and the line with most support is deemed the robust fit. The points within the threshold distance are the inliers (and constitute the eponymous consensus set). The intuition is that if one of the points is an outlier then the line will not gain much support.Furthermore, scoring a line by its support has the additional advantage of favouring better fits. More generally, we wish to fit a model, in this case a line, to data, and the random sample consists of a minimal subset of the data, in this case two points, sufficient to determine the model. 


```python
def ransac(points_2d , points_3d , iters = 20): 
    g_residual = "inf"
    g_M = None
    g_p2 = None
    g_pro = None
    temp = np.concatenate((points_2d , points_3d), axis = 1)
    for i  in range(iters):
        np.random.shuffle(temp)
        p2  = temp[:6,[0,1]]
        p3  = temp[:6,[2,3,4]]

        M = sc.calculate_projection_matrix(p2, p3)
        
        projected_2d_pts, residual = evaluate_points(M, points_2d, points_3d)

        if residual < g_residual:
            g_residual = residual
            g_M = M
            g_p2 = points_2d
            g_pro = projected_2d_pts
    return g_pro , g_residual , g_M
    
```

```python
ransca_points , ransac_resid , ransac_M= ransac(points_2d , points_3d)   
print("camera Matrix")
print(M)
print('The ransac residual is {:f}'.format(ransac_resid))
```

# Calculting camera Internal and External Parameters  matrix

<b> Camera Interal calibration matrix K</b> can be found as folllow:<br>
camera matrix M $$  M = [ KR | Kt ] $$
$$ (KR)(KR)^T = KK^T $$
$ KK^T $can be solved for K using cholesky method .
<br><br>
<b>Rotation Matrix R</b> can be solved using : <br>
$$  R = MK^{-1} $$
<br><br>
<b>Transform matrix t </b> can be solved using:<br>
$$ t = K^{-1}m_4 $$
$m_4$ is the fourth column of camera matrix M.

```python
def parameters(mat):
    p4 = mat[: , [3]]
    mat = mat[ : , [0,1,2]]
    n_mat = np.dot(mat, mat.T)
    K = np.linalg.cholesky(n_mat)
    K = K.T/K[-1,-1]
    Ki = np.linalg.inv(K)
    R = np.dot(Ki, mat)
    t = np.dot(Ki, p4)
    return K ,R , t
```

```python
K , R , t = parameters(M)
print("camera matrix is : ")
print(M)
print("\n Internal camera calibration matrix(K)")
print(K)
print("\n Rotation matrix(R)")
print(R)
print("\n Transform Matrix(t)")
print(t)
print("\n")
# b = np.dot(K, R)   
# b = np.concatenate((b, np.dot(K, t)),  axis = 1)
# print(b)
```

# Zhang Method for camera calibration 

The technique proposed by Zhang only requires the camera to observe a planar pattern shown at a few (at least two) different orientations. In the camera’s model used we consider the intrinsic parameters, the extrinsic 
parameters and the radial lens distortion. This method proposes an analytic solution to solve these parameters and then create the best approximation using a minimization method.


```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)

objp = np.zeros((6*8,3), np.float32)
objp[:,:2] = np.mgrid[0:8,0:6].T.reshape(-1,2)

objpoints = [] 
imgpoints = [] 
images = []


images = ["IMG_"+str(i)+".JPG" for i in range(5456,5471)]


for fname in images:
    img = cv2.imread("data1/" + fname)
    img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)

    gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
    
    ret, corners = cv2.findChessboardCorners(gray, (8,6),None)

    if ret == True:
        objpoints.append(objp)
        corners2 = cv2.cornerSubPix(gray,corners,(11,11),(-1,-1),criteria)
        imgpoints.append(corners2)
```

```python
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera( objpoints,imgpoints, gray.shape[::-1],None,None)
```

### Creating wire Frame  

```python
rotVec,_ = cv2.Rodrigues(rvecs[0])
projection = []
temp = np.column_stack((rotVec ,tvecs[0]))
P = np.dot(mtx,temp) # CAMERA PROJECTION MATRIX
for i in range(len(objpoints[0])):
    x = np.append(objpoints[0][i] , 1)
    i_p = np.dot(np.array(P), np.array(x).T)
    i_p = (i_p/i_p[2]).astype('int')
    projection.append(i_p[:2])
```

```python
img = cv2.imread('data1/IMG_5456.JPG',cv2.IMREAD_COLOR)
img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)
prev_point = map(int ,projection[0])
cv2.circle(img , tuple(prev_point), 1, (0,255,0),5)
for i  in range(1,len(projection)):
    pts = [int(j) for j in projection[i]]
    cv2.line(img,tuple(prev_point),tuple(pts),(255,0,0),1)
    cv2.circle(img , tuple(pts), 1, (0,255,0),5)
    prev_point = pts
plt.imshow(img)
plt.title("wireframe using the projected points")
plt.show()
```

```python
import numpy as np
import cv2
import glob
```

```python
criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 30, 0.001)
cbcol = 8
cbrow = 6
```

```python
objp = np.zeros((cbrow*cbcol, 3), np.float32)
objp[:,:2] = np.mgrid[0:cbcol,0:cbrow].T.reshape(-1,2)
```

```python
objpoints = []
imgpoints = []
# images = []
# for i in range(1,15):
#     if i < 10 :
#         images.append("left0" + str(i) + ".jpg")
#         images.append("right0" + str(i) + ".jpg")
#     else : 
#         images.append("left" + str(i) + ".jpg")
#         images.append("right" + str(i) + ".jpg")
# images.remove("left10.jpg")
# images.remove("right10.jpg")
    
images = ["IMG_"+str(i)+".JPG" for i in range(5456,5471)]
```

```python
for fname in images:
    print(fname)
    img = cv2.imread("data1/" + fname)
    img = cv2.resize(img, (0,0),fx = 0.08 , fy = 0.08)
#     print("cp0")
#     print(img.shape)
    gray = cv2.cvtColor(img , cv2.COLOR_BGR2GRAY)
#     cv2.imshow('img', img)
#     cv2.waitKey()
#     print("Cp1")
    ret ,corners = cv2.findChessboardCorners(img,(cbcol,cbrow) , None)
#     print(ret)
#     print(corners)
    if ret == True:
        objpoints.append(objp)
        corners2 = cv2.cornerSubPix(gray,corners,(11,11), (-1,-1), criteria)
        imgpoints.append(corners2)
#         print("cp2")
        img = cv2.drawChessboardCorners(img, (cbcol,cbrow) , corners, ret)
        cv2.imshow('img', img)
        cv2.waitKey(5000)
    cv2.destroyAllWindows()
```

```python
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, gray.shape[::-1],None,None)
```

```python
img = cv2.imread('data/left12.jpg')
h,  w = img.shape[:2]
newcameramtx, roi=cv2.getOptimalNewCameraMatrix(mtx,dist,(w,h),1,(w,h))
print(newcameramtx)
```

```python
dst = cv2.undistort(img, mtx, dist, None, newcameramtx)

# crop the image
x,y,w,h = roi
dst = dst[y:y+h, x:x+w]
cv2.imwrite('calibresult.png',dst)
```

```python
mean_error = 0
for i in xrange(len(objpoints)):
    imgpoints2, _ = cv2.projectPoints(objpoints[i], rvecs[i], tvecs[i], mtx, dist)
    error = cv2.norm(imgpoints[i],imgpoints2, cv2.NORM_L2)/len(imgpoints2)
    mean_error += error

print "mean error: ", mean_error/len(objpoints)
```
